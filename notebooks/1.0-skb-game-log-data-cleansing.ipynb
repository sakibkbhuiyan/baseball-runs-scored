```python
## Create an empty DataFrame to store all game log data 
df_game_log_all = pd.DataFrame()

## Directory containing the text files
directory = "../data/raw"

# Regular expression pattern to match numbers at the end of filenames
pattern = r"\d+$"

## Loop through each file in the directory
## Pattern to match numbers at the end of filenames
for filename in os.listdir(directory):
    if filename.endswith(".txt") and filename[-8:-4].isdigit():
        ## Read the file into a DataFrame
        filepath = os.path.join(directory, filename)
        df = pd.read_csv(filepath, header=None, names=column_names)
        
        ## Append the DataFrame to the master DataFrame
        df_game_log_all = df_game_log_all.append(df, ignore_index=True)

print(df_game_log_all.shape)
display(df_game_log_all.head())
```

# Data Cleaning

```python
df_game_log_all['Date'] = pd.to_datetime(df_game_log_all['Date'], format='%Y%m%d')
df_game_log_all['Year'] = df_game_log_all['Date'].dt.year

## Categorize years into baseball eras
bins = [float('-inf'), 1900, 1920, 1942, 1961, 1977, 1994, 2004, pd.Timestamp.today().year]
labels = ['Pre-1900', 'Dead Ball era', 'Live Ball era', 'Integration era', 'Expansion era', 'Free Agent era', 'Steroid era', 'Contemporary era']

# Create the categorical variable based on the 'Year' column
df_game_log_all['Era'] = pd.cut(df_game_log_all['Year'], bins=bins, labels=labels, right=False)

display(df_game_log_all.head())
```

```python
selected_columns = ['Year', 'Date', 'Era', 'Visiting_Team_Score', 'Home_Team_Score']
df_game_log_all = df_game_log_all[selected_columns]

display(df_game_log_all.head())
```

```python
## Export as CSV
filepath = "../data/processed/game_log_all.csv"
df_game_log_all.to_csv(filepath, index = False)
```
